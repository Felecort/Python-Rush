```python
from matplotlib import pyplot as plt
from collections import Counter
from random import randrange
```

```python
y = [i * i for i in range(1, 20)]
x = [i for i in range(1, 20)]
plt.plot(x, y, color="green", marker="o", linestyle="solid")
plt.title("i^2")
plt.ylabel("y")
plt.xlabel("x")
plt.show()
```

```python
order = ["first", "second", "third", "fourth", "fifth"]
num_of_order = [5, 11, 17, 4, 7]
xs = [i + 0.1 for i, _ in enumerate(order)]
plt.bar(xs, num_of_order)
plt.ylabel("y")
plt.xlabel("x")
plt.xticks([i + 0.1 for i, _ in enumerate(order)], order)
plt.show()
```

```python
grades = [83, 95, 91, 87, 70, 0, 85, 82, 100, 67, 73, 77,  0]
decile = lambda grade: grade // 10 * 10
histogram = Counter(decile(grade) for grade in grades)
plt.bar([x - 4 for x in histogram.keys()], #   4
        histogram.values(),                #  
        8)                                 #  

plt.axis([-5, 105, 0, 5])
plt.xticks([10 * i for i in range(11)])
plt.xlabel("")
plt.ylabel(" ")
plt.title("   ")
plt.show()
```

```python
mentions = [500, 505]
years = [2013, 2014]
plt.bar([2013, 2014], #   ( ) 
        mentions,
        0.8)
plt.xticks(years)
plt.ylabel("   ")
# plt.ticklabel_format(useOffset=False) #      0  1 
                                      #  +23 ()
plt.axis([2012.5, 2014.5, 499, 506])
plt.title("  ")
plt.show()
```

```python
mentions = [500, 505]
years = [2013, 2014]
plt.bar(years, #   ( ) 
        mentions,
        0.8)
plt.xticks(years)
plt.ylabel("   ")
plt.ticklabel_format(useOffset=False) #      0  1 
                                      #  +23 ()
plt.axis([2012.5, 2014.5, 0, 506])
plt.title("   ")
plt.show()
```

```python
variance = [2 ** i for i in range(8)] # 
bias_squared = variance[::-1]         #  
total_error = [x + y for x, y in zip(variance, bias_squared)]
xs = [i for i, _ in enumerate(variance)]
plt.plot(xs, variance, "g-", label="")
plt.plot(xs, bias_squared, "r-.", label="^2")
plt.plot(xs, total_error, "b:", label=" ")
plt.legend(loc=9) #  label
plt.xlabel("")
plt.title("    ")
plt.show()
```

```python
friends = [randrange(40, 100) for _ in range(10)]
minutes = [randrange(100, 200) for _ in range(10)]
labels = [a for a in list("abcdefghig")]
plt.scatter(friends, minutes)
for label, friends_count, minute_count in zip(labels, friends, minutes):
    plt.annotate(label, 
                xy=(friends_count, minute_count),
                xytext=(5, -5),
                textcoords="offset points")
plt.title("      ")
plt.xlabel(" ")
plt.ylabel(",    , ")
plt.show()
```
